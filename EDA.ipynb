Có thể thấy giới tính không có nhiều tác động đến sự nghỉ việc

```python
def gender_dist_by_col(col_name = 'DailyRate'):
    female = df[df['Gender'] == 'Female'][col_name]
    male = df[df['Gender'] == 'Male'][col_name]
    
    plt.figure(figsize=(12, 6))
    plt.subplot(1,2,1)
    # Female distribution
    sns.kdeplot(female, fill=True, color='pink', label='Female')
    plt.axvline(x=female.mean(), linestyle='--', color='red', label='Mean Female ' + col_name)

    # Male age distribution
    sns.kdeplot(male, fill=True, color='lightblue', label='Male')
    plt.axvline(x=male.mean(), linestyle='--', color='blue', label='Mean Male ' + col_name)

    plt.title(col_name + ' Distribution by Gender')
    plt.xlabel(col_name)
    plt.ylabel('Density')
    plt.legend()

    plt.subplot(1,2,2)
    female = df[df['Gender'] == 'Female']
    male = df[df['Gender'] == 'Male']
    sns.histplot(female, x=col_name, kde=True, color='red', label='Female')
    sns.histplot(male, x=col_name, kde=True, color='lightblue', label='Male')
    plt.title(col_name + ' Distribution by Gender')
    plt.xlabel(col_name)
    plt.ylabel('Count')
    plt.legend()
    plt.show()
    
```

#### 3.1.4. Mức độ hài lòng

```python
def pie_plot_by_gender(col_name = 'JobSatisfaction'):
    female = df[df['Gender'] == 'Female']
    male = df[df['Gender'] == 'Male']
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.pie(female[col_name].value_counts().values, labels=female[col_name].value_counts().keys(),  autopct='%.0f%%')
    plt.title(f'{col_name} in Female')
    plt.subplot(1,2,2)
    plt.pie(male[col_name].value_counts().values, labels=male[col_name].value_counts().keys(),  autopct='%.0f%%')
    plt.title(f'{col_name} in Male')
    plt.show()
```

```python
pie_plot_by_gender()
```

```python
def attrition_count_by_gender(col_name='JobSatisfaction', percentage=False):
    plt.figure(figsize=(12, 6))
    
    for i, gender in enumerate(['Female', 'Male'], 1):
        plt.subplot(1, 2, i)
        df_gender = df[df['Gender'] == gender]
        attrition_counts = df_gender.groupby(col_name)['Attrition'].count()
        
        if percentage:
            attrition_counts = (attrition_counts / attrition_counts.sum()) * 100
            ylabel = 'Attrition Percentage'
        else:
            ylabel = 'Attrition Count'
        
        sns.barplot(
            x=attrition_counts.index, 
            y=attrition_counts.values, 
            palette='tab10'
        )
        
        plt.ylabel(ylabel)
        plt.title(f'Attrition in {gender}')
    
    plt.tight_layout()
    plt.show()
```

```python
attrition_count_by_gender(percentage=True)
```

* Có thể thấy mặc dù hài lòng với công việc, nhưng nhóm 3 và 4 vẫn có xu hướng rời bỏ công ty nhiều hơn

#### 3.2.1. Theo số năm làm việc tại công ty

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, y='MonthlyRate', x='YearsAtCompany', hue='Gender')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
def working_rank(x):
    if x < 5:
        return 'Under 5 years'
    elif x <= 10:
        return '5 - 10 years'
    elif x <= 20:
        return '10 - 20 years'
    else:
        return 'Over 20 years'

df['WorkingYearsRank'] = df['TotalWorkingYears'].apply(working_rank)
df['CompanyWorkingYearsRank'] = df['YearsAtCompany'].apply(working_rank)
```

```python
monthly_rate_by_working_years = df.groupby('CompanyWorkingYearsRank').MonthlyRate.mean().reset_index().sort_index()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data=monthly_rate_by_working_years,
            x = 'CompanyWorkingYearsRank',
            y = 'MonthlyRate',
            palette='tab10')
plt.xlabel('Years At Company')

plt.subplot(1,2,2)
attr_by_working_years = df.groupby('CompanyWorkingYearsRank').Attrition.count().reset_index().sort_index()
sns.barplot(data=attr_by_working_years,
            x = 'CompanyWorkingYearsRank',
            y = 'Attrition',
            palette='tab10')
plt.xlabel('Years At Company')
plt.ylabel('Attrition Count')
```

* Mức lương giữa theo số năm làm việc không có quá nhiều chênh lệch giữa các nhóm

* Mặc dù vậy, các nhân viên lâu năm (trên 10 năm) có xu hướng ở lại công ty, trong khi các nhân viên làm việc tại công ty dưới 10 năm lại có xu hướng rời đi nhiều hơn
